## The Iris Dataset

**What is it?**
- 150 iris flower samples
- 3 species: Setosa, Versicolor, Virginica
- 4 measurements per flower:
  - Sepal length (cm)
  - Sepal width (cm)
  - Petal length (cm)
  - Petal width (cm)

**Goal:** Predict species based on measurements

---

## Step 1: Import Libraries

First, let's import all the Python libraries we need.

````python
# Data manipulation
import pandas as pd
import numpy as np

# Machine Learning
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✅ All libraries imported successfully!")
print(f"Python version: {pd.__version__}")
````

## Step 2: Load the Dataset

The Iris dataset comes built-in with scikit-learn!

````python
# Load the Iris dataset
iris = load_iris()

# Extract features (X) and target (y)
X = iris.data  # Measurements (150 samples × 4 features)
y = iris.target  # Species labels (0, 1, or 2)

# Get feature names and target names
feature_names = iris.feature_names
target_names = iris.target_names

print("✅ Dataset loaded successfully!")
print(f"\nNumber of samples: {len(X)}")
print(f"Number of features: {X.shape[1]}")
print(f"Number of classes: {len(target_names)}")
print(f"\nFeature names: {feature_names}")
print(f"Target names (species): {list(target_names)}")
````

## Step 3: Explore the Data

Let's convert to pandas DataFrame for easier exploration.

````python
# Create DataFrame
df = pd.DataFrame(X, columns=feature_names)
df['species'] = y
df['species_name'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Display first 10 rows
print("First 10 samples:")
df.head(10)
````

````python
# Statistical summary
print("Statistical Summary:")
df.describe()
````

````python
# Check class distribution
print("Class Distribution:")
print(df['species_name'].value_counts())
print("\n✅ Perfectly balanced dataset (50 samples per class)!")
````

## Step 4: Visualize the Data

Let's create some beautiful visualizations!

````python
# Pairplot - shows relationships between all features
plt.figure(figsize=(12, 10))
sns.pairplot(df, hue='species_name', markers=['o', 's', 'D'])
plt.suptitle('Iris Dataset - Feature Relationships', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()

print("✅ Pairplot created!")
print("\nNotice how different species cluster together?")
print("This is why ML can learn to classify them!")
````

````python
# Box plots for each feature
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Feature Distributions by Species', fontsize=16, y=1.02)

for idx, feature in enumerate(feature_names):
    row = idx // 2
    col = idx % 2
    
    sns.boxplot(data=df, x='species_name', y=feature, ax=axes[row, col])
    axes[row, col].set_title(f'{feature}')
    axes[row, col].set_xlabel('Species')
    axes[row, col].set_ylabel(feature)

plt.tight_layout()
plt.show()

print("✅ Box plots created!")
print("\nObservation: Petal measurements seem more distinctive than sepal measurements.")
````

## Step 5: Prepare Data for Machine Learning

Split data into training set (80%) and testing set (20%).

````python
# Split the data
# X = features (measurements)
# y = target (species)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% for testing
    random_state=42,    # For reproducibility
    stratify=y          # Keep class balance in both sets
)

print("✅ Data split completed!")
print(f"\nTraining set: {len(X_train)} samples")
print(f"Testing set: {len(X_test)} samples")
print(f"\nTraining classes distribution:")
unique, counts = np.unique(y_train, return_counts=True)
for cls, count in zip(unique, counts):
    print(f"  {target_names[cls]}: {count} samples")
````

## Step 6: Train the Machine Learning Model

We'll use a **Decision Tree Classifier** - a simple but powerful ML algorithm.

````python
# Create the model
model = DecisionTreeClassifier(
    random_state=42,
    max_depth=3  # Limit depth to avoid overfitting
)

print("Model created!")
print(f"Model type: {type(model).__name__}")
print(f"\nTraining the model...")

# Train the model
model.fit(X_train, y_train)

print("\n✅ MODEL TRAINED SUCCESSFULLY!")
print("\nThe model has learned patterns from 120 flower samples.")
print("Now let's see how well it can predict species for flowers it has never seen!")
````

## Step 7: Make Predictions

Let's use our trained model to predict species for the test set.

````python
# Predict on training set
y_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)

# Predict on test set
y_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("✅ Predictions completed!")
print(f"\n📊 RESULTS:")
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
print(f"Testing Accuracy: {test_accuracy * 100:.2f}%")
print(f"\n🎉 WOW! The model correctly classifies {test_accuracy * 100:.1f}% of flowers!")
````

````python
# Show some predictions vs actual
print("\nSample Predictions (first 10 test samples):")
print("="*60)
print(f"{'Actual':<15} {'Predicted':<15} {'Correct?':<10}")
print("="*60)

for i in range(min(10, len(y_test))):
    actual = target_names[y_test[i]]
    predicted = target_names[y_test_pred[i]]
    correct = '✅' if y_test[i] == y_test_pred[i] else '❌'
    print(f"{actual:<15} {predicted:<15} {correct:<10}")
````

## Step 8: Detailed Evaluation

Let's see detailed metrics and create a confusion matrix.

````python
# Classification report
print("Detailed Classification Report:")
print("="*60)
print(classification_report(y_test, y_test_pred, target_names=target_names))

print("\nWhat these metrics mean:")
print("  • Precision: Of all flowers predicted as X, how many were actually X?")
print("  • Recall: Of all actual X flowers, how many did we find?")
print("  • F1-Score: Balance between precision and recall")
print("  • Support: Number of samples for each class")
````

````python
# Confusion Matrix
cm = confusion_matrix(y_test, y_test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names,
            yticklabels=target_names,
            cbar_kws={'label': 'Count'})
plt.title('Confusion Matrix - Iris Classification', fontsize=14, fontweight='bold')
plt.ylabel('Actual Species', fontsize=12)
plt.xlabel('Predicted Species', fontsize=12)
plt.tight_layout()
plt.show()

print("\n✅ Confusion matrix created!")
print("\nHow to read:")
print("  • Diagonal = Correct predictions")
print("  • Off-diagonal = Mistakes")
print("  • Darker blue = More samples")
````

## Step 9: Feature Importance

Which measurements are most important for classification?

````python
# Get feature importance from the trained model
feature_importance = model.feature_importances_

# Create DataFrame for easy visualization
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importance
}).sort_values('Importance', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
bars = plt.barh(importance_df['Feature'], importance_df['Importance'])

# Color bars
colors = plt.cm.viridis(importance_df['Importance'] / importance_df['Importance'].max())
for bar, color in zip(bars, colors):
    bar.set_color(color)

plt.xlabel('Importance Score', fontsize=12)
plt.title('Feature Importance in Iris Classification', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n📊 Feature Importance Rankings:")
for idx, row in importance_df.iterrows():
    print(f"  {row['Feature']:<25} {row['Importance']:.4f}")

print(f"\n✅ Most important feature: {importance_df.iloc[0]['Feature']}")
````

## Step 10: Try Your Own Prediction!

Let's predict the species of a new flower with custom measurements.

````python
# Define a new flower (change these values to experiment!)
new_flower = np.array([[
    5.1,  # sepal length (cm)
    3.5,  # sepal width (cm)
    1.4,  # petal length (cm)
    0.2   # petal width (cm)
]])

print("New Flower Measurements:")
print("="*40)
for i, feature in enumerate(feature_names):
    print(f"{feature:<25} {new_flower[0][i]:.1f} cm")

# Make prediction
prediction = model.predict(new_flower)[0]
predicted_species = target_names[prediction]

# Get prediction probabilities (requires different approach for Decision Tree)
probabilities = model.predict_proba(new_flower)[0]

print("\n" + "="*40)
print(f"🌸 PREDICTED SPECIES: {predicted_species.upper()}")
print("="*40)

print("\nPrediction Confidence:")
for i, species in enumerate(target_names):
    print(f"  {species:<15} {probabilities[i]*100:>6.2f}%")

print("\n✅ Try changing the measurements above and re-run to see different predictions!")
````

## Step 11: Visualize Decision Boundaries (Advanced)

Let's visualize how the model makes decisions (using 2 features for simplicity).

````python
# Use only 2 features for visualization (petal length and width)
X_2d = X[:, [2, 3]]  # Petal length and width

# Train a new model with just these 2 features
model_2d = DecisionTreeClassifier(random_state=42, max_depth=3)
model_2d.fit(X_2d, y)

# Create mesh for decision boundary
h = 0.02  # Step size
x_min, x_max = X_2d[:, 0].min() - 0.5, X_2d[:, 0].max() + 0.5
y_min, y_max = X_2d[:, 1].min() - 0.5, X_2d[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict for each point in mesh
Z = model_2d.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.figure(figsize=(12, 8))
plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.RdYlBu)
scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y, 
                     s=100, edgecolor='k', alpha=0.8, cmap=plt.cm.RdYlBu)
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)
plt.title('Decision Tree Classification - Decision Boundaries', 
         fontsize=14, fontweight='bold')
plt.colorbar(scatter, label='Species')

# Add legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=plt.cm.RdYlBu(i/2), label=target_names[i]) 
                   for i in range(3)]
plt.legend(handles=legend_elements, loc='upper left')

plt.tight_layout()
plt.show()

print("\n✅ Decision boundary visualization created!")
print("\nWhat you see:")
print("  • Colored regions = Model's predictions")
print("  • Points = Actual flowers")
print("  • Boundaries = Where model changes predictions")
````
