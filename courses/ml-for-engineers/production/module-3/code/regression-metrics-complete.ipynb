## Part 1: Understanding Regression Metrics

### The 4 Main Metrics

1. **MAE (Mean Absolute Error)**
   - Average absolute difference between predictions and actual
   - Easy to interpret (same units as target)
   - Formula: $\frac{1}{n}\sum|y_i - \hat{y}_i|$

2. **MSE (Mean Squared Error)**
   - Average of squared differences
   - Penalizes large errors more
   - Formula: $\frac{1}{n}\sum(y_i - \hat{y}_i)^2$

3. **RMSE (Root Mean Squared Error)**
   - Square root of MSE
   - Same units as target (like MAE)
   - More sensitive to outliers than MAE
   - Formula: $\sqrt{\frac{1}{n}\sum(y_i - \hat{y}_i)^2}$

4. **R² (R-squared / Coefficient of Determination)**
   - Proportion of variance explained by model
   - Range: 0 to 1 (higher is better)
   - 1.0 = perfect fit, 0.0 = no better than mean
   - Formula: $1 - \frac{SS_{res}}{SS_{tot}}$

```python
# Create example predictions
np.random.seed(42)

# True values
y_true = np.array([100, 120, 115, 130, 125, 140, 135, 150, 145, 160])

# Good predictions (close to true)
y_pred_good = y_true + np.random.normal(0, 3, len(y_true))

# Poor predictions (far from true)
y_pred_poor = y_true + np.random.normal(0, 15, len(y_true))

# Perfect predictions
y_pred_perfect = y_true.copy()

print('Example Data:')
print(f'True values: {y_true}')
print(f'\nGood predictions: {y_pred_good.round(2)}')
print(f'Poor predictions: {y_pred_poor.round(2)}')
```

### Calculate All Metrics

```python
def calculate_all_metrics(y_true, y_pred, model_name='Model'):
    """Calculate and display all regression metrics"""
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    print(f'{model_name} Metrics:')
    print(f'  MAE:  {mae:.2f}')
    print(f'  MSE:  {mse:.2f}')
    print(f'  RMSE: {rmse:.2f}')
    print(f'  R²:   {r2:.4f}')
    print()
    
    return {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2': r2}

# Compare models
metrics_perfect = calculate_all_metrics(y_true, y_pred_perfect, 'Perfect Model')
metrics_good = calculate_all_metrics(y_true, y_pred_good, 'Good Model')
metrics_poor = calculate_all_metrics(y_true, y_pred_poor, 'Poor Model')
```

### Interpretation Guide

- **MAE = $3**: On average, predictions are off by $3
- **RMSE = $5**: Typical error is $5 (emphasizes larger errors)
- **R² = 0.85**: Model explains 85% of variance (good!)
- **R² = 0.50**: Model explains 50% of variance (moderate)
- **R² < 0.30**: Model is weak
- **R² < 0**: Model is worse than just predicting the mean!
