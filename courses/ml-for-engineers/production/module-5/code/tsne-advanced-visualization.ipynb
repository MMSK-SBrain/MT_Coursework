# t-SNE: Advanced Dimensionality Reduction

## Session 5: t-SNE vs PCA for Visualization

**Objectives:** Master t-SNE, compare to PCA, understand when to use each

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import time
import warnings
warnings.filterwarnings('ignore')
print('Ready!')
```

## Load MNIST Digits

```python
digits = load_digits()
X = digits.data
y = digits.target
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(f'Data shape: {X.shape}')
```

## Method 1: PCA (Fast, Linear)

```python
start = time.time()
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
time_pca = time.time() - start
print(f'PCA time: {time_pca:.2f}s')
```

## Method 2: t-SNE (Slow, Non-linear)

```python
start = time.time()
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_tsne = tsne.fit_transform(X_scaled)
time_tsne = time.time() - start
print(f't-SNE time: {time_tsne:.2f}s')
print(f't-SNE is {time_tsne/time_pca:.1f}x slower than PCA')
```

## Side-by-Side Comparison

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# PCA
for digit in range(10):
    mask = y == digit
    ax1.scatter(X_pca[mask, 0], X_pca[mask, 1], label=str(digit), alpha=0.6, s=20)
ax1.set_title(f'PCA (Time: {time_pca:.2f}s)', fontsize=13, fontweight='bold')
ax1.legend(title='Digit', bbox_to_anchor=(1.05, 1), loc='upper left')
ax1.grid(True, alpha=0.3)

# t-SNE
for digit in range(10):
    mask = y == digit
    ax2.scatter(X_tsne[mask, 0], X_tsne[mask, 1], label=str(digit), alpha=0.6, s=20)
ax2.set_title(f't-SNE (Time: {time_tsne:.2f}s)', fontsize=13, fontweight='bold')
ax2.legend(title='Digit', bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.grid(True, alpha=0.3)

plt.suptitle('PCA vs t-SNE: MNIST Digits Visualization', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Perplexity Comparison

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
perplexities = [5, 30, 50]

for idx, perp in enumerate(perplexities):
    tsne_temp = TSNE(n_components=2, random_state=42, perplexity=perp, n_iter=500)
    X_tsne_temp = tsne_temp.fit_transform(X_scaled[:500])  # Subset for speed
    
    scatter = axes[idx].scatter(X_tsne_temp[:, 0], X_tsne_temp[:, 1], 
                               c=y[:500], cmap='tab10', alpha=0.6, s=20)
    axes[idx].set_title(f'Perplexity = {perp}', fontsize=12, fontweight='bold')
    axes[idx].grid(True, alpha=0.3)

plt.suptitle('Effect of Perplexity on t-SNE', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## When to Use PCA vs t-SNE?

**Use PCA when:**
- Speed is important
- Need to transform new data
- Want interpretable components
- Global structure matters
- Preprocessing for other algorithms

**Use t-SNE when:**
- Visualization is primary goal
- Local structure matters
- Have computational resources
- Don't need to transform new data
- Beautiful plots desired!

**Recommendation:** Use PCA first, then t-SNE for final visualization!
