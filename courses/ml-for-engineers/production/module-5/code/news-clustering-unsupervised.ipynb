# News Article Clustering with NLP

## Session 3: Document Clustering and Topic Discovery

**Objectives:** Apply clustering to text data, discover topics without labels

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD
from wordcloud import WordCloud
import warnings
warnings.filterwarnings('ignore')
print('Ready!')
```

## Generate Synthetic News Articles

```python
# Synthetic news articles
sports_articles = [
    'football soccer goal team player match win championship',
    'basketball game score NBA finals team player championship',
    'tennis match player tournament grand slam win',
] * 100

tech_articles = [
    'artificial intelligence machine learning AI technology data',
    'smartphone mobile app technology innovation software',
    'cybersecurity hacking data breach security technology',
] * 100

business_articles = [
    'stock market investment trading finance economy',
    'company earnings profit revenue business growth',
    'startup venture capital funding investment business',
] * 100

articles = sports_articles + tech_articles + business_articles
true_labels = ['Sports']*300 + ['Tech']*300 + ['Business']*300

df = pd.DataFrame({'text': articles, 'true_category': true_labels})
print(f'Total articles: {len(df)}')
```

## TF-IDF Vectorization

```python
vectorizer = TfidfVectorizer(max_features=100, stop_words='english')
X_tfidf = vectorizer.fit_transform(df['text'])
print(f'TF-IDF matrix shape: {X_tfidf.shape}')
```

## K-Means Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X_tfidf)
print('\nCluster distribution:')
print(df['cluster'].value_counts())
```

## Topic Discovery - Top Terms per Cluster

```python
print('\nTop terms per cluster:')
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()

for i in range(3):
    print(f'\nCluster {i}:')
    top_terms = [terms[ind] for ind in order_centroids[i, :10]]
    print(', '.join(top_terms))
```

## Visualization with t-SNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_tfidf.toarray())

plt.figure(figsize=(12, 8))
for i in range(3):
    mask = df['cluster'] == i
    plt.scatter(X_tsne[mask, 0], X_tsne[mask, 1], label=f'Cluster {i}', alpha=0.6, s=30)
plt.title('News Articles Clustered (t-SNE Visualization)', fontsize=14, fontweight='bold')
plt.legend()
plt.tight_layout()
plt.show()
```

## Summary

**Key Learnings:**
- TF-IDF converts text to numerical features
- K-Means successfully groups similar articles
- Topics emerge from top terms in each cluster
- t-SNE helps visualize document clusters
