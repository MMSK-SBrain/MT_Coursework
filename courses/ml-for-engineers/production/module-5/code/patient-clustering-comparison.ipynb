# Patient Clustering - Hierarchical & DBSCAN Comparison

## Session 2: Comparing Clustering Algorithms on Healthcare Data

**Objectives:** Compare K-Means, Hierarchical, and DBSCAN clustering on patient medical data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
import warnings
warnings.filterwarnings('ignore')
print('Ready!')
```

## Generate Synthetic Patient Data

```python
np.random.seed(42)
n_patients = 800

# Generate patient features
patients = pd.DataFrame({
    'age': np.random.randint(18, 85, n_patients),
    'bmi': np.random.uniform(18, 40, n_patients),
    'blood_pressure_sys': np.random.randint(90, 180, n_patients),
    'blood_pressure_dia': np.random.randint(60, 120, n_patients),
    'glucose': np.random.uniform(70, 200, n_patients),
    'cholesterol': np.random.uniform(150, 300, n_patients),
    'heart_rate': np.random.randint(60, 100, n_patients)
})

print(f'Generated {len(patients)} patients')
patients.head()
```

## Method 1: K-Means Clustering

```python
# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(patients)

# K-Means with k=4
kmeans = KMeans(n_clusters=4, random_state=42)
patients['cluster_kmeans'] = kmeans.fit_predict(X_scaled)
silhouette_kmeans = silhouette_score(X_scaled, patients['cluster_kmeans'])
print(f'K-Means Silhouette Score: {silhouette_kmeans:.4f}')
```

## Method 2: Hierarchical Clustering

```python
# Create dendrogram
linkage_matrix = linkage(X_scaled[:200], method='ward')  # Subset for visualization
plt.figure(figsize=(14, 6))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram', fontsize=14, fontweight='bold')
plt.xlabel('Patient Index')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```

```python
# Hierarchical clustering
hierarchical = AgglomerativeClustering(n_clusters=4, linkage='ward')
patients['cluster_hierarchical'] = hierarchical.fit_predict(X_scaled)
silhouette_hier = silhouette_score(X_scaled, patients['cluster_hierarchical'])
print(f'Hierarchical Silhouette Score: {silhouette_hier:.4f}')
```

## Method 3: DBSCAN Clustering

```python
# DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=10)
patients['cluster_dbscan'] = dbscan.fit_predict(X_scaled)
print(f'DBSCAN found {len(set(patients["cluster_dbscan"])) - (1 if -1 in patients["cluster_dbscan"] else 0)} clusters')
print(f'Outliers: {(patients["cluster_dbscan"] == -1).sum()} patients')
```

## Comparison and Insights

```python
print('\nClustering Method Comparison:')
print('='*60)
print(f'K-Means: {len(set(patients["cluster_kmeans"]))} clusters, Silhouette={silhouette_kmeans:.4f}')
print(f'Hierarchical: {len(set(patients["cluster_hierarchical"]))} clusters, Silhouette={silhouette_hier:.4f}')
print(f'DBSCAN: {len(set(patients["cluster_dbscan"])) - 1} clusters + outliers')
print('='*60)
print('\nRecommendation: Use K-Means or Hierarchical for this dataset')
print('DBSCAN useful for outlier detection in patient data')
```

## Summary

**Key Learnings:**
- K-Means: Fast, requires K
- Hierarchical: Dendrogram useful, computationally expensive
- DBSCAN: Finds outliers, no K needed
- Choose based on data characteristics and business needs
