## 5. Data Selection and Filtering

### 5.1 Selecting Single Column

```python
# Select only the closing prices
closing_prices = df['Adj Close']

print("First 10 closing prices:")
print(closing_prices.head(10))
```

### 5.2 Selecting Multiple Columns

```python
# Select Date, Close, and Volume columns
subset = df[['Date', 'Adj Close', 'Volume']]

print("Subset with 3 columns:")
print(subset.head(10))
```

### 5.3 Filtering Rows by Condition

```python
# Find days where closing price was above $200
high_price_days = df[df['Adj Close'] > 200]

print(f"Days with price above $200: {len(high_price_days):,}")
print(f"That's {len(high_price_days)/len(df)*100:.1f}% of all trading days")
print("\nFirst 10 such days:")
print(high_price_days.head(10))
```

```python
# Find days with very high volume (> 100 million shares)
high_volume_days = df[df['Volume'] > 100_000_000]

print(f"Days with volume > 100M shares: {len(high_volume_days):,}")
print("\nFirst 10 such days:")
print(high_volume_days[['Date', 'Adj Close', 'Volume']].head(10))
```

### 5.4 Multiple Conditions

```python
# Find days with high price AND high volume
high_price_and_volume = df[
    (df['Adj Close'] > 200) & 
    (df['Volume'] > 100_000_000)
]

print(f"Days with price > $200 AND volume > 100M: {len(high_price_and_volume):,}")
print("\nSample of such days:")
print(high_price_and_volume[['Date', 'Adj Close', 'Volume']].head(10))
```

## 6. Creating New Columns (Feature Engineering)

### 6.1 Daily Price Change

```python
# Calculate daily price change (in dollars)
df['Price_Change'] = df['Adj Close'].diff()

print("Daily Price Change:")
print(df[['Date', 'Adj Close', 'Price_Change']].head(10))
```

### 6.2 Daily Return (Percentage Change)

```python
# Calculate daily return (percentage change)
df['Daily_Return'] = df['Adj Close'].pct_change() * 100

print("Daily Returns (%):")
print(df[['Date', 'Adj Close', 'Daily_Return']].head(10))

print("\nDaily Return Statistics:")
print(f"Average daily return: {df['Daily_Return'].mean():.2f}%")
print(f"Volatility (std dev): {df['Daily_Return'].std():.2f}%")
```

### 6.3 Moving Average

```python
# Calculate 10-day moving average
df['MA_10'] = df['Adj Close'].rolling(window=10).mean()

# Calculate 50-day moving average
df['MA_50'] = df['Adj Close'].rolling(window=50).mean()

print("Price with Moving Averages:")
print(df[['Date', 'Adj Close', 'MA_10', 'MA_50']].tail(10))
```

### 6.4 Price Range

```python
# Calculate daily price range (High - Low)
df['Price_Range'] = df['High'] - df['Low']

print("Daily Price Range:")
print(df[['Date', 'High', 'Low', 'Price_Range']].head(10))

print(f"\nAverage daily price range: ${df['Price_Range'].mean():.2f}")
```

## 7. Finding Specific Days

### 7.2 Lowest Price Day

```python
# Find the day with lowest closing price
min_price_idx = df['Adj Close'].idxmin()
min_price_day = df.loc[min_price_idx]

print("Day with Lowest Closing Price:")
print("="*80)
print(f"Date: {min_price_day['Date']}")
print(f"Price: ${min_price_day['Adj Close']:.2f}")
print(f"Volume: {min_price_day['Volume']:,.0f} shares")
```

### 7.3 Highest Volume Day

```python
# Find the day with highest trading volume
max_volume_idx = df['Volume'].idxmax()
max_volume_day = df.loc[max_volume_idx]

print("Day with Highest Trading Volume:")
print("="*80)
print(f"Date: {max_volume_day['Date']}")
print(f"Volume: {max_volume_day['Volume']:,.0f} shares")
print(f"Price: ${max_volume_day['Adj Close']:.2f}")
```

### 7.4 Biggest Gain Day

```python
# Find day with biggest percentage gain
max_return_idx = df['Daily_Return'].idxmax()
max_return_day = df.loc[max_return_idx]

print("Day with Biggest Gain:")
print("="*80)
print(f"Date: {max_return_day['Date']}")
print(f"Return: {max_return_day['Daily_Return']:.2f}%")
print(f"Price: ${max_return_day['Adj Close']:.2f}")
```

### 7.5 Biggest Loss Day

```python
# Find day with biggest percentage loss
min_return_idx = df['Daily_Return'].idxmin()
min_return_day = df.loc[min_return_idx]

print("Day with Biggest Loss:")
print("="*80)
print(f"Date: {min_return_day['Date']}")
print(f"Return: {min_return_day['Daily_Return']:.2f}%")
print(f"Price: ${min_return_day['Adj Close']:.2f}")
```

## 8. Grouping and Aggregation

### 8.1 Convert Date to Datetime

```python
# Convert Date column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Extract year and month
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Year_Month'] = df['Date'].dt.to_period('M')

print("Date components extracted:")
print(df[['Date', 'Year', 'Month', 'Year_Month']].head())
```

### 8.2 Yearly Statistics

```python
# Group by year and calculate statistics
yearly_stats = df.groupby('Year')['Adj Close'].agg([
    ('Start_Price', 'first'),
    ('End_Price', 'last'),
    ('Min_Price', 'min'),
    ('Max_Price', 'max'),
    ('Avg_Price', 'mean'),
    ('Total_Days', 'count')
]).round(2)

# Calculate yearly return
yearly_stats['Yearly_Return_%'] = (
    (yearly_stats['End_Price'] - yearly_stats['Start_Price']) / 
    yearly_stats['Start_Price'] * 100
).round(2)

print("Yearly Statistics:")
print("="*80)
print(yearly_stats)
```

### 8.3 Monthly Average Volume

```python
# Group by year-month and calculate average volume
monthly_volume = df.groupby('Year_Month')['Volume'].mean().round(0)

print("Average Monthly Trading Volume:")
print("="*80)
print(monthly_volume.tail(12))  # Last 12 months
```

## 10. Comparing Multiple Stocks

### 10.1 Load Multiple Stocks

```python
# Load multiple stocks
stocks = {}
tickers = ['AAPL', 'GOOGL', 'MSFT', 'TSLA']

for ticker in tickers:
    filepath = f'../datasets/{ticker}_5y_SAMPLE.csv'
    stocks[ticker] = pd.read_csv(filepath)
    stocks[ticker]['Date'] = pd.to_datetime(stocks[ticker]['Date'])
    print(f"✓ Loaded {ticker}: {len(stocks[ticker])} days")
```

### 10.2 Stock Comparison Table

```python
# Create comparison table
comparison = []

for ticker, data in stocks.items():
    stats = {
        'Ticker': ticker,
        'Avg_Price': data['Adj Close'].mean(),
        'Min_Price': data['Adj Close'].min(),
        'Max_Price': data['Adj Close'].max(),
        'Volatility': data['Adj Close'].std(),
        'Avg_Volume': data['Volume'].mean(),
        'Total_Days': len(data)
    }
    comparison.append(stats)

comparison_df = pd.DataFrame(comparison)

print("Stock Comparison:")
print("="*80)
print(comparison_df.round(2))
```

### 10.3 Total Returns Comparison

```python
# Calculate total return for each stock
returns = []

for ticker, data in stocks.items():
    start_price = data['Adj Close'].iloc[0]
    end_price = data['Adj Close'].iloc[-1]
    total_return = (end_price - start_price) / start_price * 100

    returns.append({
        'Ticker': ticker,
        'Start_Price': start_price,
        'End_Price': end_price,
        'Total_Return_%': total_return
    })

returns_df = pd.DataFrame(returns)
returns_df = returns_df.sort_values('Total_Return_%', ascending=False)

print("Total Returns Comparison (5 years):")
print("="*80)
print(returns_df.round(2))
```

```python
print("="*80)
print("APPLE (AAPL) STOCK ANALYSIS SUMMARY")
print("="*80)
print(f"\nData Period: {df['Date'].min().date()} to {df['Date'].max().date()}")
print(f"Total Trading Days: {len(df):,}")

print("\n📊 PRICE STATISTICS")
print("-" * 80)
print(f"Average Price:        ${df['Adj Close'].mean():.2f}")
print(f"Starting Price:       ${df['Adj Close'].iloc[0]:.2f}")
print(f"Ending Price:         ${df['Adj Close'].iloc[-1]:.2f}")
print(f"Minimum Price:        ${df['Adj Close'].min():.2f}")
print(f"Maximum Price:        ${df['Adj Close'].max():.2f}")
print(f"Price Range:          ${df['Adj Close'].max() - df['Adj Close'].min():.2f}")

total_return = (df['Adj Close'].iloc[-1] - df['Adj Close'].iloc[0]) / df['Adj Close'].iloc[0] * 100
print(f"\nTotal Return:         {total_return:.2f}%")

print("\n📈 VOLATILITY")
print("-" * 80)
print(f"Price Std Dev:        ${df['Adj Close'].std():.2f}")
print(f"Daily Return Std Dev: {df['Daily_Return'].std():.2f}%")
print(f"Average Daily Return: {df['Daily_Return'].mean():.2f}%")

print("\n📊 VOLUME")
print("-" * 80)
print(f"Average Volume:       {df['Volume'].mean():,.0f} shares")
print(f"Median Volume:        {df['Volume'].median():,.0f} shares")
print(f"Max Volume:           {df['Volume'].max():,.0f} shares")

print("\n🎯 KEY DAYS")
print("-" * 80)
print(f"Highest Price:        ${max_price_day['Adj Close']:.2f} on {max_price_day['Date'].date()}")
print(f"Lowest Price:         ${min_price_day['Adj Close']:.2f} on {min_price_day['Date'].date()}")
print(f"Biggest Gain:         {max_return_day['Daily_Return']:.2f}% on {max_return_day['Date'].date()}")
print(f"Biggest Loss:         {min_return_day['Daily_Return']:.2f}% on {min_return_day['Date'].date()}")

print("\n✓ DATA QUALITY")
print("-" * 80)
print(f"Missing Values:       {df.isnull().sum().sum()}")
print(f"Duplicate Dates:      {df['Date'].duplicated().sum()}")
print(f"Data Quality:         {'✓ GOOD' if df.isnull().sum().sum() == 0 else '⚠ NEEDS REVIEW'}")

print("\n" + "="*80)
print("Analysis Complete!")
print("="*80)
```
