## 1. Setup and Imports

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Set figure size default
plt.rcParams['figure.figsize'] = (14, 7)
plt.rcParams['font.size'] = 11

print("✓ Libraries imported successfully")
print(f"matplotlib version: {plt.matplotlib.__version__}")
print(f"seaborn version: {sns.__version__}")
print(f"plotly version: {go.__version__}")
```

## 2. Load Stock Data

```python
# Load Apple stock data
df = pd.read_csv('../datasets/AAPL_5y_SAMPLE.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)

# Calculate additional features
df['Daily_Return'] = df['Adj Close'].pct_change() * 100
df['MA_20'] = df['Adj Close'].rolling(window=20).mean()
df['MA_50'] = df['Adj Close'].rolling(window=50).mean()
df['MA_200'] = df['Adj Close'].rolling(window=200).mean()

print(f"✓ Loaded {len(df)} days of data")
print(f"Date range: {df['Date'].min().date()} to {df['Date'].max().date()}")
df.head()
```

## 3. Line Charts

### 3.1 Simple Line Chart - Closing Price

```python
# Create simple line chart
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Adj Close'], color='#2E86DE', linewidth=1.5)
plt.title('Apple (AAPL) Stock Price - 5 Year History', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("📊 This chart shows the overall price trend over 5 years")
```

### 3.2 Line Chart with Moving Averages

```python
# Chart with multiple moving averages
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Adj Close'], label='Price', color='#2E86DE', linewidth=2, alpha=0.8)
plt.plot(df['Date'], df['MA_20'], label='20-Day MA', color='#FF6B6B', linewidth=1.5, alpha=0.7)
plt.plot(df['Date'], df['MA_50'], label='50-Day MA', color='#4ECDC4', linewidth=1.5, alpha=0.7)
plt.plot(df['Date'], df['MA_200'], label='200-Day MA', color='#FFA502', linewidth=1.5, alpha=0.7)

plt.title('AAPL Price with Moving Averages', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("📊 Moving averages smooth out price fluctuations")
print("   - Short MA (20-day) follows price closely")
print("   - Long MA (200-day) shows long-term trend")
```

### 3.3 Recent Price Action (Last 6 Months)

```python
# Filter last 6 months (approximately 125 trading days)
recent = df.tail(125)

plt.figure(figsize=(14, 7))
plt.plot(recent['Date'], recent['Adj Close'], color='#2E86DE', linewidth=2, marker='o', markersize=2)
plt.fill_between(recent['Date'], recent['Adj Close'], alpha=0.2, color='#2E86DE')

plt.title('AAPL Price - Last 6 Months', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 4. Bar Charts

### 4.1 Volume Bar Chart

```python
# Volume bar chart (last 100 days)
recent = df.tail(100)

plt.figure(figsize=(14, 7))
colors = ['#EE5A6F' if ret < 0 else '#4ECDC4' for ret in recent['Daily_Return']]
plt.bar(recent['Date'], recent['Volume'], color=colors, alpha=0.6, width=0.8)

plt.title('AAPL Trading Volume - Last 100 Days', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Volume (Shares)', fontsize=12)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print("📊 Red bars = price down, Teal bars = price up")
print(f"Average volume: {recent['Volume'].mean():,.0f} shares")
```

### 4.2 Monthly Average Volume

```python
# Calculate monthly average volume
df['Year_Month'] = df['Date'].dt.to_period('M')
monthly_volume = df.groupby('Year_Month')['Volume'].mean().tail(24)

plt.figure(figsize=(14, 7))
plt.bar(range(len(monthly_volume)), monthly_volume.values, color='#4ECDC4', alpha=0.7)
plt.xticks(range(len(monthly_volume)), monthly_volume.index.astype(str), rotation=45, ha='right')

plt.title('AAPL Average Monthly Volume - Last 2 Years', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average Daily Volume', fontsize=12)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

## 5. Histograms

### 5.1 Price Distribution

```python
# Price distribution histogram
plt.figure(figsize=(14, 7))
plt.hist(df['Adj Close'], bins=50, color='#4ECDC4', alpha=0.7, edgecolor='black')

# Add mean line
mean_price = df['Adj Close'].mean()
plt.axvline(mean_price, color='red', linestyle='--', linewidth=2, label=f'Mean: ${mean_price:.2f}')

plt.title('AAPL Price Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Price (USD)', fontsize=12)
plt.ylabel('Frequency (Number of Days)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print("📊 This shows which price ranges were most common")
```

### 5.2 Daily Returns Distribution

```python
# Daily returns histogram
returns = df['Daily_Return'].dropna()

plt.figure(figsize=(14, 7))
plt.hist(returns, bins=60, color='#4ECDC4', alpha=0.7, edgecolor='black')

# Add mean and zero lines
plt.axvline(0, color='black', linestyle='-', linewidth=1, label='Zero')
plt.axvline(returns.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {returns.mean():.2f}%')

plt.title('AAPL Daily Returns Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Daily Return (%)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print(f"Mean return: {returns.mean():.2f}%")
print(f"Std dev (volatility): {returns.std():.2f}%")
print(f"Best day: +{returns.max():.2f}%")
print(f"Worst day: {returns.min():.2f}%")
```

## 6. Scatter Plots

### 6.1 Price vs Volume

```python
# Scatter plot: Price vs Volume
plt.figure(figsize=(12, 8))
plt.scatter(df['Adj Close'], df['Volume']/1e6, alpha=0.5, c=range(len(df)), cmap='viridis')

plt.title('AAPL: Price vs Trading Volume', fontsize=16, fontweight='bold')
plt.xlabel('Price (USD)', fontsize=12)
plt.ylabel('Volume (Millions of Shares)', fontsize=12)
plt.colorbar(label='Time (older → newer)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Calculate correlation
correlation = df['Adj Close'].corr(df['Volume'])
print(f"Correlation between price and volume: {correlation:.3f}")
```

### 6.2 Returns vs Volume

```python
# Scatter: Daily Returns vs Volume
plt.figure(figsize=(12, 8))
colors = ['#EE5A6F' if ret < 0 else '#4ECDC4' for ret in df['Daily_Return']]
plt.scatter(df['Daily_Return'], df['Volume']/1e6, alpha=0.4, c=colors, s=30)

plt.title('Daily Returns vs Trading Volume', fontsize=16, fontweight='bold')
plt.xlabel('Daily Return (%)', fontsize=12)
plt.ylabel('Volume (Millions)', fontsize=12)
plt.axvline(0, color='black', linestyle='--', linewidth=1)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. Candlestick Charts (OHLC)

### 7.1 Interactive Candlestick Chart with Plotly

```python
# Create candlestick chart for last 100 days
recent = df.tail(100)

fig = go.Figure(data=[go.Candlestick(
    x=recent['Date'],
    open=recent['Open'],
    high=recent['High'],
    low=recent['Low'],
    close=recent['Close'],
    name='AAPL'
)])

fig.update_layout(
    title='AAPL Candlestick Chart - Last 100 Days',
    yaxis_title='Price (USD)',
    xaxis_title='Date',
    height=600,
    template='plotly_white'
)

fig.show()

print("📊 Green candles = price up, Red candles = price down")
print("   Hover over candles to see OHLC values")
```

### 7.2 Candlestick with Volume

```python
# Candlestick with volume subplot
recent = df.tail(100)

fig = make_subplots(
    rows=2, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.05,
    subplot_titles=('AAPL Price', 'Volume'),
    row_heights=[0.7, 0.3]
)

# Candlestick
fig.add_trace(
    go.Candlestick(
        x=recent['Date'],
        open=recent['Open'],
        high=recent['High'],
        low=recent['Low'],
        close=recent['Close'],
        name='AAPL'
    ),
    row=1, col=1
)

# Volume bars
colors = ['red' if row['Close'] < row['Open'] else 'green' for idx, row in recent.iterrows()]
fig.add_trace(
    go.Bar(x=recent['Date'], y=recent['Volume'], marker_color=colors, name='Volume'),
    row=2, col=1
)

fig.update_layout(
    title='AAPL: Price and Volume',
    height=700,
    template='plotly_white',
    showlegend=False
)

fig.update_yaxes(title_text="Price (USD)", row=1, col=1)
fig.update_yaxes(title_text="Volume", row=2, col=1)
fig.update_xaxes(title_text="Date", row=2, col=1)

fig.show()
```

## 8. Multi-Stock Comparison

### 8.1 Load Multiple Stocks

```python
# Load multiple stocks
tickers = ['AAPL', 'GOOGL', 'MSFT', 'TSLA']
stocks = {}

for ticker in tickers:
    stocks[ticker] = pd.read_csv(f'../datasets/{ticker}_5y_SAMPLE.csv')
    stocks[ticker]['Date'] = pd.to_datetime(stocks[ticker]['Date'])
    print(f"✓ Loaded {ticker}")
```

### 8.2 Normalized Price Comparison

```python
# Normalize prices to start at 100
plt.figure(figsize=(14, 8))

colors = {'AAPL': '#2E86DE', 'GOOGL': '#EE5A6F', 'MSFT': '#4ECDC4', 'TSLA': '#FFA502'}

for ticker, data in stocks.items():
    normalized = (data['Adj Close'] / data['Adj Close'].iloc[0]) * 100
    plt.plot(data['Date'], normalized, label=ticker, linewidth=2, color=colors[ticker])

plt.title('Tech Stocks Comparison - Normalized to 100', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Normalized Price (Start = 100)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("📊 All stocks start at 100 for easy comparison")
print("   Higher line = better performance")
```

### 8.3 Side-by-Side Subplots

```python
# Create 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Tech Stocks - Individual Charts', fontsize=18, fontweight='bold')

for idx, (ticker, data) in enumerate(stocks.items()):
    row = idx // 2
    col = idx % 2

    axes[row, col].plot(data['Date'], data['Adj Close'], color=colors[ticker], linewidth=2)
    axes[row, col].set_title(f'{ticker}', fontsize=14, fontweight='bold')
    axes[row, col].set_xlabel('Date', fontsize=11)
    axes[row, col].set_ylabel('Price (USD)', fontsize=11)
    axes[row, col].grid(True, alpha=0.3)

    # Add stats text
    avg_price = data['Adj Close'].mean()
    axes[row, col].text(
        0.02, 0.98, f'Avg: ${avg_price:.2f}',
        transform=axes[row, col].transAxes,
        fontsize=10,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    )

plt.tight_layout()
plt.show()
```

## 9. Heatmap Correlation

### 9.1 Daily Returns Correlation

```python
# Create DataFrame with all stock returns
returns_df = pd.DataFrame()

for ticker, data in stocks.items():
    data['Daily_Return'] = data['Adj Close'].pct_change() * 100
    returns_df[ticker] = data['Daily_Return']

# Calculate correlation matrix
corr_matrix = returns_df.corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt='.3f',
    cmap='RdYlGn',
    center=0,
    square=True,
    linewidths=1,
    cbar_kws={"shrink": 0.8}
)

plt.title('Stock Returns Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("📊 Higher correlation = stocks move together")
print("   1.0 = perfect correlation")
print("   0.0 = no correlation")
```

## 10. Advanced: Interactive Dashboard with Plotly

### 10.1 Complete Interactive Dashboard

```python
# Load data with all features
df = stocks['AAPL'].copy()
df['MA_20'] = df['Adj Close'].rolling(20).mean()
df['MA_50'] = df['Adj Close'].rolling(50).mean()

# Create subplot figure
fig = make_subplots(
    rows=3, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.05,
    subplot_titles=('Price with Moving Averages', 'Volume', 'Daily Returns'),
    row_heights=[0.5, 0.25, 0.25]
)

# Price chart with MAs
fig.add_trace(
    go.Scatter(x=df['Date'], y=df['Adj Close'], name='Price', line=dict(color='#2E86DE', width=2)),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=df['Date'], y=df['MA_20'], name='20-Day MA', line=dict(color='#FF6B6B', width=1)),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=df['Date'], y=df['MA_50'], name='50-Day MA', line=dict(color='#4ECDC4', width=1)),
    row=1, col=1
)

# Volume
colors_vol = ['red' if df.loc[idx, 'Adj Close'] < df.loc[max(0, idx-1), 'Adj Close'] else 'green' 
              for idx in df.index]
fig.add_trace(
    go.Bar(x=df['Date'], y=df['Volume'], name='Volume', marker_color=colors_vol),
    row=2, col=1
)

# Daily Returns
df['Daily_Return'] = df['Adj Close'].pct_change() * 100
colors_ret = ['red' if ret < 0 else 'green' for ret in df['Daily_Return']]
fig.add_trace(
    go.Bar(x=df['Date'], y=df['Daily_Return'], name='Daily Return', marker_color=colors_ret),
    row=3, col=1
)

# Update layout
fig.update_layout(
    title_text='AAPL Stock Analysis Dashboard',
    height=900,
    template='plotly_white',
    showlegend=True,
    hovermode='x unified'
)

fig.update_yaxes(title_text="Price (USD)", row=1, col=1)
fig.update_yaxes(title_text="Volume", row=2, col=1)
fig.update_yaxes(title_text="Return (%)", row=3, col=1)
fig.update_xaxes(title_text="Date", row=3, col=1)

fig.show()

print("📊 Interactive Dashboard Created!")
print("   - Hover to see details")
print("   - Zoom and pan enabled")
print("   - Click legend to toggle series")
```
