3. Fine-tuning (unfreeze top layers)

````python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping

print(f'TensorFlow: {tf.__version__}')
````

````python
# Load pre-trained VGG16 (without top classification layer)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze base model
base_model.trainable = False

print('VGG16 loaded and frozen')
print(f'Total layers: {len(base_model.layers)}')
````

````python
# Build complete model
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(5, activation='softmax')  # 5 classes for Flowers dataset
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
````

## Key Insights

### Transfer Learning Advantages:
1. **Faster Training:** Minutes instead of hours
2. **Better Accuracy:** Pre-trained features are powerful
3. **Less Data Needed:** Works with small datasets
4. **Lower Computational Cost:** Don't need to train from scratch

### When to Use Transfer Learning:
- Small datasets (<10k images)
- Similar domain (ImageNet → other images)
- Limited compute resources
- Need quick results

### Best Practices:
1. Start with feature extraction (freeze base)
2. Train classification head
3. Then fine-tune (unfreeze top layers)
4. Use very small learning rate for fine-tuning (1e-5)
5. Apply data augmentation
